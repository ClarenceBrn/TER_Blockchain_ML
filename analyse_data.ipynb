```python
import pandas as pd

df = pd.read_csv('active_2024_addresses.csv')

print(f"Taille initiale du dataset : {len(df)} lignes")
print(df.head())

nb_entites_par_classe = df.groupby('category')['wallet'].nunique()
print("--- Nombre d'entités (wallets) par classe ---")
print(nb_entites_par_classe)
print("\n--- Répartition en % ---")
print((nb_entites_par_classe / nb_entites_par_classe.sum()) * 100)

# On garde seulement les entites avec plus d'une addresse
df_filtered = df.groupby('wallet').filter(lambda x: len(x) > 1)

print(f"Taille après filtrage : {len(df_filtered)} lignes")

# Compter combien d'entités uniques il reste
unique_entities = df_filtered['wallet'].nunique()
print(f"\nNombre d'entités restantes : {unique_entities}")

# Sauvegarder le résultat pour la suite du pipeline
df_filtered.to_csv('dataset_filtered.csv', index=False)
```

```python
nb_entites_par_classe = df_filtered.groupby('category')['wallet'].nunique()

print("--- Nombre d'entités (wallets) par classe ---")
print(nb_entites_par_classe)
print("\n--- Répartition en % ---")
print((nb_entites_par_classe / nb_entites_par_classe.sum()) * 100)
```
